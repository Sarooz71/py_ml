```python
import pandas as pd

# Day 3
# Creating a DataFrame from a dictionary

data = {
    'student_name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'hours_studied': [5, 8, 3, 10, 6],
    'Previous_score': [65, 85, 55, 92, 75],
    'attendance': [80, 90, 70, 95, 85],
    'sleep_hours': [7, 8, 6, 9, 7],
    'extracurricular': ['yes', 'no', 'yes', 'yes', 'no'],
    'Parent_education': ['Bachelor', 'Master', 'Highschool', 'PhD', 'Bachelor']
}

df = pd.DataFrame(data)

print("Student Performance DataFrame:")
print(df)

print("Data info")
df.info()

#basic dataframe infromation 
print("\nDataframe shape :",df.shape[0],"rows",df.shape[1],"columns")
print("column names:",df.columns.tolist())
print("Data types:\n",df.dtypes)

#view data 

print("first 3 rows:")
print(df.head(3))

print("last 2 rows:")
print(df.tail(2))

#statistical memory
print("\n statistical memory") #mean median mode
print(df.describe())

#check for missing values
print("\n Missing values:")
print(df.isnull().sum())

#information about dataframe
print("\n Dataframe info:")
df.info()

# Check NaN values
print("\nCheck for NaN values:")
print(df.isnull())

#check on duplicates rows
print("\n Duplicate ")
```
